```python
utf_chars = ({r'\\xc3\\xa1':r'á',r'\\xc3\\xa9':r'é',r'\\xc3\\xad':r'í',r'\\xc3\\xb3':r'ó',
             r'\\xc3\\xba':r'ú',r'\\xc3\\xb1':r'ñ',r'\\xef\\xac\\x81':r'fi',
             r'\\xe2\\x80\\x9c':r'"',r'\\xe2\\x80\\x9d':r'"',r'\\xef\\xac\\x82':r'fl',
             r'\\xc2\\xb0':r'',r'\\xc3\\x93':r'Ó',r'\\xc2\\xba':r'',r'\\xc3\\x91':r'Ñ',
             r'\\xe2\\x80\\x93':r'-',r'\\xc3\\x81':r'Á',r'\\xc3\\x8d':r'Í',r'\\xc2\\xa1':r'',
             r'\\xef\\x82\\xb7':r'','\xc2\xab':r'','\xc2\xbb':r'',r'\\x0c':r'',r'\\xe2\\x80\\xa6': r'...',
             r'(?<=\w)([\\]n)((?=[a-z]))':r' '})
def conv_chars(string, chars=utf_chars):
    for utf, char in chars.items():
        string=re.sub(utf, char, string)
    return string
def prep_for_stanza(string):
    string = conv_chars(re.sub(r'http\S+', '', string))
    return re.sub(r'\\n', r' ', re.sub(r'\\n\\n', r'\n\n', string))
def convert_stanza(df):
    docs = [nlp(prep_for_stanza(text)) for text in df.text]
    docs = [[sent for sent in doc.sentences if len(sent.text)>20] for doc in docs]
    if len(docs)!=len(df.text):
        return print('ERROR')
    return docs

#nlp = stanza.Pipeline(lang='es', processors='tokenize,mwt,pos,lemma,depparse,ner', tokenize_no_ssplit=True)
```

```python
nmf = NMF(15)
nmf0 = nmf.fit_transform(X)
```

```python
def get_top_terms(topic, n_terms, nmf=nmf, terms=cv.get_feature_names_out()):
    # get the topic components (i.e., term weights)
    comps = nmf.components_[topic]
    # get term indices, sorted (descending) by topic weights
    top_indices = np.argsort(comps)[-n_terms:]
    # use the `terms` array to get the actual top terms
    top_terms = np.array(terms)[top_indices]
    return top_terms
```
